```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from scipy.stats import norm
```

```python
x= np.linspace(-6,6,1000)
```

```python
fx=norm.pdf(x,loc=0,scale=1)  #loc-mean and scale=Std deviation. It's 0,1 for a normal distribution
```

```python
plt.plot(x,fx);
```

```python
Fx=norm.cdf(x,loc=0,scale=1)
```

```python
plt.plot(x,Fx);
```

```python
logfx=norm.logpdf(x,loc=0,scale=1)
plt.plot(x,logfx);
```

```python
logcfx=norm.logcdf(x,loc=0,scale=1)
plt.plot(x,logcfx);
```

```python
x=np.linspace(-6,6,50)
```

```python
fx=norm.pdf(x,loc=0,scale=1)
```

```python
filt=np.outer(fx,fx)
```

```python
plt.imshow(filt,cmap='gray');
```

```python
from scipy.signal import convolve2d
```

```python
out=convolve2d(gray,filt)
```

```python
plt.subplot(1,2,1)
plt.imshow(gray,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(out,cmap='gray');
```
