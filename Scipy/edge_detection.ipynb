```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
from PIL import Image
from scipy.signal import convolve2d
```

```python
Hx=[[1,0,-1],[2,0,-2],[1,0,-1]]
```

```python
Hy=[[1,2,1],[0,0,0],[-1,-2,-1]]
```

```python
!wget https://avatars.githubusercontent.com/u/20688701?v=4
```

```python
im=Image.open('20688701?v=4')
```

```python
A=np.mean(im,axis=2)
```

```python
plt.imshow(A)
```

```python
plt.imshow(Hx);
```

```python
plt.imshow(Hy);
```

```python
Gx=convolve2d(A,Hx)
Gy=convolve2d(A,Hy)
```

```python
plt.imshow(Gx);
```

```python
plt.imshow(Gy);
```

```python
G=np.sqrt((np.power(Gx,2))+(np.power(Gy,2)))
```

```python
plt.imshow(G);
```

```python
plt.subplot(1,2,1)
plt.imshow(A,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(G,cmap='gray');
```
