**Scatter Plot**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.random.randn(100,2)
plt.scatter(X[:,0],X[:,1])
```

```python
X=np.random.randn(200,2)
X[:50]+=3
Y=np.zeros(200)
Y[:50]=1
```

```python
plt.scatter(X[:,0],X[:,1],c=Y);
```
