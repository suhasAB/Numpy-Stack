```python
import numpy as np
import matplotlib.pyplot as plt
x=(np.random.random(1000)*2)-1
y=(np.random.random(1000)*2)-1
```

```python
z=x*y
x.shape,y.shape,z.shape
L=[]
```

```python
for i in z:
  if i>0:
    L.append(0)
  else:
    L.append(1)


```

```python
z=np.array(L)
z.shape
```

```python
plt.scatter(x,y,c=z);
```
